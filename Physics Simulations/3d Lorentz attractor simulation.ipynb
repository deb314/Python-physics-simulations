```python
%matplotlib qt
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
import numpy as np
from mpl_toolkits import mplot3d

h=0.01  #discreteness parameter

class Lorentz():
    def __init__(self,sigma,b,r,x,y,z,iters):
        self.sigma=sigma
        self.b=b
        self.r=r
        self.x=x
        self.y=y
        self.z=z
        self.iters=iters
        self.x1=np.zeros((self.iters,1))
        self.y1=np.zeros((self.iters,1))
        self.z1=np.zeros((self.iters,1))
        self.time=np.zeros((self.iters,1))
        
    def deriv(self):
        h=0.01
        t=0.0
        for i in range(0,self.iters):
            self.x=self.x+h*self.sigma*(self.y-self.x)
            self.y=self.y+h*(self.r*self.x-self.y-self.x*self.z)
            self.z=self.z+h*(self.x*self.y-self.b*self.z)
            
            self.x1[i,0]=self.x
            self.y1[i,0]=self.y
            self.z1[i,0]=self.z
            t=t+h
            self.time[i,0]=t
            
iter_arg=2000
            
lor1=Lorentz(10,8/3,28,8,10,0,iter_arg)
lor1.deriv()

lor2=Lorentz(10,8/3,28,15,1,1,iter_arg)
lor2.deriv()

lor3=Lorentz(10,8/3,28,5,0,10,iter_arg)
lor3.deriv()

still=[]
for i in range(iter_arg-1):
    still.append(i)
#-------------------------------------------------------------------------------------
    
#animation for the actual 3d lorentz attractor
    
xline1=[]
yline1=[]
zline1=[]

xline3=[]
yline3=[]
zline3=[]
    
xline2=[]
yline2=[]
zline2=[]

def animate(i):
    xline1.append(lor1.x1[i,0])
    yline1.append(lor1.y1[i,0])
    zline1.append(lor1.z1[i,0])
    line1.set_data(xline1,yline1)
    line1.set_3d_properties(zline1)
    dot1.set_data(lor1.x1[i,0],lor1.y1[i,0])
    dot1.set_3d_properties(lor1.z1[i,0])
    
    xline2.append(lor2.x1[i,0])
    yline2.append(lor2.y1[i,0])
    zline2.append(lor2.z1[i,0])
    line2.set_data(xline2,yline2)
    line2.set_3d_properties(zline2)
    dot2.set_data(lor2.x1[i,0],lor2.y1[i,0])
    dot2.set_3d_properties(lor2.z1[i,0])
    
    xline3.append(lor3.x1[i,0])
    yline3.append(lor3.y1[i,0])
    zline3.append(lor3.z1[i,0])
    line3.set_data(xline3,yline3)
    line3.set_3d_properties(zline3)
    dot3.set_data(lor3.x1[i,0],lor3.y1[i,0])
    dot3.set_3d_properties(lor3.z1[i,0])
    
    ax.view_init(30,30*i*0.04)
    page.canvas.draw()
    return line1,line2,line3,dot1,dot2,dot3,

#--------------------------------------------------------------------------------------------------

#drawing the octants

xpo=np.zeros((10,10))
ypo=np.zeros((10,10))
zpo=np.zeros((10,10))
xpo1=np.zeros((10,10))
ypo1=np.zeros((10,10))
zpo1=np.zeros((10,10))

for i in range(0,10):
    for j in range(0,10):
        xpo[i,j]=-30+6*j
        ypo[i,j]=-30+6*i
        
for i in range(0,10):
    for j in range(0,10):
        zpo[i,j]=0
        
for i in range(0,10):
    for j in range(0,10):
        zpo1[i,j]=0+(55/10)*i
        ypo1[i,j]=-30+(60/10)*j
        xpo1[i,j]=0
#--------------------------------------------------------------------------------------------------------
        
#animating the time vs position part

liney=[]
linet=[]

def animate2(i):
    linet.append(lor1.time[i,0])
    liney.append(lor1.y1[i,0])
    line4.set_data(linet,liney)
    return line4,
#----------------------------------------------------------------------------------------------------
    
#creating the page for the actual lorentz attractor simulation

page=plt.figure(figsize=(9,9))
ax=page.add_axes([0,0,1,1], projection="3d")
ax.set_xlim(-30,30)
ax.set_ylim(-30,30)
ax.set_zlim(0,55)
ax.set_axis_off()
line1,=ax.plot([],[],[],'r',linewidth=2)
dot1,=ax.plot([],[],[],'ro')

line2,=ax.plot([],[],[],'g',linewidth=2)
dot2,=ax.plot([],[],[],'go')

line3,=ax.plot([],[],[],'m',linewidth=2)
dot3,=ax.plot([],[],[],'mo')

ax.plot_surface(xpo,ypo,zpo,color='blue' ,alpha=0.2)
ax.plot_surface(xpo1,ypo1,zpo1,color='blue',alpha=0.2)
ax.plot_surface(ypo1,xpo1,zpo1,color='blue',alpha=0.2)
#--------------------------------------------------------------------------------------------------------

#creating a page for the time vs coordinates simulation

page2=plt.figure('x-coordinate vs time plot',figsize=(7,3))
ax2=page2.add_axes([0.1,0.025,0.8,0.95])
ax2.set_ylim(-25,25)
ax2.set_xlim(0,50)
ax2.set_title('y-position vs time for the red one')
ax2.text(1,20,'Initial value of x=10')
ax2.spines['bottom'].set_position('center')
ax2.spines['top'].set_visible(False)
ax2.spines['right'].set_visible(False)
ax2.set_xlabel('time',)
ax2.set_ylabel('x position')

line4,=ax2.plot([],[],'r',linewidth=2)
#---------------------------------------------------------------------------------------------------------

#the animation part

anim=FuncAnimation(page,animate,frames=still,interval=h*500,blit=True,repeat=False)
anim2=FuncAnimation(page2,animate2,frames=still, interval=h*500, blit=True,repeat=False)
plt.show()
    
    
```
